# Сборный проект. Анализ данных по продажам игр.

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

Необходимо рассмотреть данные для прогнозирования на 2017 год.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Цель проекта: выявить определяющие успешность игры закономерности, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании на 2017 год.**

**Описание данных**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
dates = [str(i) for i in range (1980, 2017)]
def fill_year(row):
    name = row['name']
    for i in dates:
        if i in name:
            return int(i)
```

```python
df['new_year'] = df.apply(fill_year, axis=1)
```

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

*H0 - средние пользовательские рейтинги платформ Xbox One и PC одинаковые*

*H1 - средние пользовательские рейтинги платформ Xbox One и PC не одинаковые*

```python
df_xone = df_new.loc[(df_new['platform'] == 'XOne') & (df['user_score'] != 0)]
df_xone.head()
```

```python
df_pc = df_new.loc[(df_new['platform'] == 'PC') & (df['user_score'] != 0)]
df_pc.head()
```

```python
# задаём уровень значимости
alpha = 0.05

# проводим ttest для двух выборок, в качестве альтернативной используем одностороннюю гипотезу «больше»
results = st.ttest_ind(df_xone['user_score'], df_pc['user_score'])

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

```python
df_xone['user_score'].mean()
```

```python
df_pc['user_score'].mean()
```

**Есть основания утверждать, что средние пользовательские рейтинги платформ XOne и PC равны друг другу.**

#### Средние пользовательские рейтинги жанров Action и Sports разные.

*H0 - средние пользовательские рейтинги жанров Action и Sports одинаковые*

*H1 - средние пользовательские рейтинги жанров Action и Sports разные*

```python
df_action = df_new.loc[(df_new['genre'] == 'Action')  & (df['user_score'] != 0)]
df_action.head()
```

```python
df_sports = df_new.loc[(df_new['genre'] == 'Sports') & (df['user_score'] != 0)]
df_sports.head()
```

```python
# задаём уровень значимости
alpha = 0.05

# проводим ttest для двух выборок, в качестве альтернативной используем одностороннюю гипотезу «больше»
results = st.ttest_ind(df_action['user_score'], df_sports['user_score'])

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

```python
df_action['user_score'].mean()
```

```python
df_sports['user_score'].mean()
```

**Есть основания утверждать, что средние пользовательские рейтинги жанров Action и Sports не равны друг другу.**

**На основании обновленной таблицы, к потенциально прибыльным платформам следует отнести PS4, Xone, 3DS. Их срок жизни пока составляет 4-6 лет и наблюдаются высокие продажи. Для дальнейшей деятельности компании "Стримчик" рекомендуется обратить внимание на игры, взаимодействующие преимущественно с этими платформами. В частности, для платформы PS4 рекомендуется обратить внимание на след игры:  Call of Duty: Black Ops 3, Grand Theft Auto V, FIFA, Star Wars Battlefront (2015), Сall of Duty: Advanced Warfare. Варианты для игр по другим платформам можно рассмотреть подробнее в п.3.5. проекта. При этом необходимо учесть, что для пользователей Европы, а следовательно, скорее всего, и для пользователей РФ предпочтительнее игры на PS4 и XOne, нежели на 3DS, поскольку 3DS - платформа, которая пользуется наибольшей популярностью в Японии. Ввиду совершенно разных менталитетов, предполагаем, что такие игры должны быть в ассортиментном портфеле, но не более 5-10% от общего ассортимента. Мы также проверили гипотезу о равенстве пользовательских оценок данных платформ. Есть основания утверждать (но не однозначно), что пользовтельские оценки PS4 и XOne равны, поэтому делать акцент на одной платформе не стоит.**

**Дополнительно, мы рассмотрели на наличие линейной связи отзывы пользователей и критиков, и пришли к выводу, что связь есть, но слабая. Нельзя однозначно утверждать, что оценки критиков или пользователей влияют на количество продаж.**

**Самым продаваемым жанром для пользователей однозначно является жанр Shooting, далее Sports. Однако, если смотреть на распределение общих продаж, также можно увидеть, что ввиду наибольшего разнообразия игр в жанре Action, общие продажи по этому жанры самые высокие. К тому же, средняя оценка игр Action выше, чем Sports. Мы также проверили гипотезу о равенстве пользовательских оценок данных жанров (Sports и Action). Есть основания утверждать (но не однозначно), что пользовательские оценки не равны. Поэтому рекомендуется в ассортиментный портфель включать игры разных жанров.
